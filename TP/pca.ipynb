```python
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../DataSet/decathlon.txt", sep="\t")
data
```

```python
# éliminer les colonnes que nous n'utiliserons pas
my_data = data.drop(['Points', 'Rank', 'Competition'], axis=1)

# transformer les données en array numpy
X = my_data.values
print(X)
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

```python
from sklearn import decomposition

pca = decomposition.PCA(n_components=2)
pca.fit(X_scaled)
```

```python
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```

```python
# projeter X sur les composantes principales
X_projected = pca.transform(X_scaled)

# afficher chaque observation
plt.scatter(X_projected[:, 0], X_projected[:, 1],
    # colorer en utilisant la variable 'Rank'
    c=data.get('Rank'))

plt.xlim([-5.5, 5.5])
plt.ylim([-4, 4])
plt.colorbar()
```

```python
X_tp = [[0,0], [0.5, 0.5], [1, 1], [1.5, 1.5], [2, 2], [2.5, 2.5], [3, 3]]
std_scale = preprocessing.StandardScaler().fit_transform(X_tp)

pca = decomposition.PCA(n_components=2)
pca.fit(std_scale)

print(pca.components_)
print(pca.explained_variance_ratio_.sum())
```

```python
X_tp = [[1,3, 1, 2], [1, 1, 2, 0]]
std_scale = preprocessing.StandardScaler().fit_transform(X_tp)

pca = decomposition.PCA(n_components=2)
pca.fit(X_tp)

print(pca.components_)
print(pca.explained_variance_ratio_.sum())
```
